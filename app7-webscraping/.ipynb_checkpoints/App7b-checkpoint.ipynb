```python
# empty list
l = []
for item in all:
    #print(item.find_all("h5",{"":""}))
    #print(item.find("h5",{"":""}))
    #print(item.find("h5",{"":""}).text)
    # new dictionary 
    d = {}
    print(item.find_all("h5",{"":""})[0].text)
    d["Tua de"] = item.find_all("h5",{"":""})[0].text
    
    d["Gia"] = item.find("span",{"class":"price"}).text
    print(item.find("span",{"class":"location"}).text.replace("\n","").replace(" ",""))
    try:
        print(item.find("span",{"class":"create-at"}).text.replace("\n","").replace(" ",""))
        d["Ngay tao"] = item.find("span",{"class":"create-at"}).text.replace("\n","").replace(" ","")
        print(item.find("span",{"class":"create-at"}).find("b").text)
    except:
        pass
        #Sprint(None)
    try:
        print(item.find("div",{"class":"description"}).text.replace("\n","").replace(" ",""))
        d["Mieu ta"] = item.find("div",{"class":"description"}).text.replace("\n","").replace(" ","")
    except:
        #pass
        print(None)
    try:    
        for column_group in item.find_all("div",{"class":"columnGroup"}):
            print(column_group)
            for feature_group,feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):
                #print(feature_group.text,feature_name.text)
                if "Lot size" in feature_group.text:
                    print(feature_name.text)
    except:
        pass
    print(" ")
    
    l.append(d)
```
